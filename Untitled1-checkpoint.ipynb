```python
import random
# from PIL import Image

import imageio
import matplotlib.pyplot as plt
%matplotlib inline
picG = imageio.imread('gabor.png')    
plt.figure(figsize = (5,5))
plt.imshow(picG)



def stats():    
    stats = {"strength":random.randint(1,25), "Dexterity":random.randint(1,25), "Constitution":random.randint(1,25), "Intelligence":random.randint(1,25), "Wisdom":random.randint(1,25), "Charisma":random.randint(1,25)}
    return(stats)



gabor = stats()
print("Minecraft Steve")
for t in gabor:
    print(t,gabor[t])
```
